## Data Cleaning

```python
# Check for column-wise null count and indicate whether values are missing 
student.isnull().sum() # sum of the values for the requested axis = 0
```

```python
# Check for row-wise null count 
student.isnull().sum(axis=1) # axis = 1, columns
```

```python
# Check for column-wise null percentages
round(100*(student.isnull().sum()/len(student.index)), 2) #len() =  length (the number of items) of an object.
```

```python
# Drop unneccessary columns
student = student.drop('serial_no', axis=1)
student = student.drop('hobbies', axis=1)
student = student.drop('mentor_name', axis=1)

# Check for column-wise null percentages again post dropping unnecessary columns
round(100*(student.isnull().sum()/len(student.index)), 2)
```

### Now, on inspection you might notice that some columns have large percentage of Null values. Drop all the rows which have Null values for such columns.

```python
# Drop rows with high % of null values in columns with high null value %
student = student[~np.isnan(student['sports_marks'])] #Test element-wise for NaN and return result as a boolean array.

# Check for column-wise null percentages again post dropping unnecessary rows
round(100*(student.isnull().sum()/len(student.index)), 2)
```

```python
# count the number of rows having >= 2 missing values
len(student[student.isnull().sum(axis=1) >= 2].index)
```

```python
# Write your code for dropping the rows with more than 2 null values here
student = student[student.isnull().sum(axis=1) <= 2]

# Check for column-wise null percentages again post dropping unnecessary rows
round(100*(student.isnull().sum()/len(student.index)), 2)
```

```python
# Inspect the language column
student['language'].describe()
```

```python
# fill the NaN values in the 'language' column here
student["language"].fillna("English", inplace = True) 

# Check for column-wise null percentages again post dropping unnecessary rows
round(100*(student.isnull().sum()/len(student.index)), 2)
```
