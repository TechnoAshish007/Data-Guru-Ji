```python
# Read the data set using pandas
sales = pd.read_csv("emp_sales.csv")
sales.head()
```

```python
# Group by the data for each employee for sales done by him in every season
sales1 = sales.groupby(['Employee'])['Spring','Winter','Autumn', 'Summer'].sum()
sales1
```

```python
# Total Sales made the employees in the whole year (all 4 seasons combined)
sales1.sum(axis=1) # sales.groupby(['Employee'])['Spring','Winter','Autumn', 'Summer'].sum().sum(axis=1)
```

```python
# Find the emmployees which have done sales of less than 1000 across all the seasons.
sales1[sales1.sum(axis= 1)<1000]
```

```python
# Group the employees that have done sales less than 1000 across all the seasons under ‘Others’ 
sales.replace(sales1[sales1.sum(axis=1)<1000].index, 'Others', inplace=True)
sales.head(10)
```
